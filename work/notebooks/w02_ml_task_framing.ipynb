## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
"""
Task type: Scoring (a form of regression/ranking-adjacent problem)

My lane (Refresh / Content Opportunity Scoring) is best framed as a scoring task,
not classification or clustering. I'm not sorting pages into fixed categories - I'm
assigning each page a continuous "refresh opportunity" score based on multiple signals
(search volume, position, CTR gap, content age). Pages can then be ranked by this score,
which is what makes it a scoring/ranking problem rather than a binary classification
("refresh" vs "don't refresh").
"""
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
"""
Target: A continuous "refresh opportunity score" per page.

This isn't an observed outcome that already exists in the data - it's a proxy I'm
defining, built from a weighted combination of existing signals: how far the page's
CTR falls below what's typical for its position, how much search volume it commands,
and how outdated the content is (content_age_days). The label comes from a defined
rule I construct from these signals, not from a pre-existing "ground truth" column,
since we don't have historical data on which refreshes actually worked.
"""
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
"""
Success metric: Precision@K for the top N recommended pages.

If we recommend the top 50 pages by opportunity score, we'd consider it 'good' if a
meaningful share of those pages later show improved CTR or position after being
refreshed (compared to a random or lowest-scored set of 50 pages). This ties the
metric directly to the real action - a team refreshing a fixed number of pages per
sprint - rather than optimizing an abstract score with no operational meaning.
"""
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
"""
A fixed rule like "if position > 10, refresh" ignores how these signals interact.
Two pages can both sit at position 12, but one has high search volume and a large CTR
gap (genuine opportunity), while the other has near-zero volume (not worth the effort).
A single-threshold rule treats them identically. An ML/scoring approach can weigh
multiple signals together and rank pages by a combined opportunity score, capturing
patterns a simple if-statement can't - like the fact that search_volume alone barely
predicts traffic (correlation ~0.001), so volume needs to be combined with position
and CTR to mean anything.
"""
```
